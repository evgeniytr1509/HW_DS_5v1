```python
import pandas as pd
import os

#  путь к каталогу, содержащему файлы CSV
base_directory = r'f:\Dell\Data_Science\HW_DS_5\data'

# Создание словаря, в котором будут храниться датафреймы для каждой активности
activity_dataframes = {}

# чтение каждого каталога в базовом каталоге
for activity_dir in ['idle', 'running', 'stairs', 'walking']:
    directory_path = os.path.join(base_directory, activity_dir)
    
    # пустой список, в который будут добавлены DataFrame для данной активности
    activity_dataframes[activity_dir] = []

    # список файлов в указанном каталоге
    file_list = os.listdir(directory_path)

    # чтение каждого файла в DataFrame
    for file_name in file_list:
        if file_name.endswith(".csv"):  
            file_path = os.path.join(directory_path, file_name)
            df = pd.read_csv(file_path)
            activity_dataframes[activity_dir].append(df)

# словарь activity_dataframes, содержащий датафреймы для каждой активности

idle_df = pd.concat(activity_dataframes['idle'], ignore_index=True)
running_df = pd.concat(activity_dataframes['running'], ignore_index=True)
stairs_df = pd.concat(activity_dataframes['stairs'], ignore_index=True)
walking_df = pd.concat(activity_dataframes['walking'], ignore_index=True)


print("Idle DataFrame:")
print(idle_df.head(5))

print("Running DataFrame:")
print(running_df.head(5))

print("Stairs DataFrame:")
print(stairs_df.head(5))

print("Walking DataFrame:")
print(walking_df.head(5))
```

```python
import os

# путь каталогу "data"
data_directory = r'f:\Dell\Data_Science\HW_DS_5\data'

# пустой словарь действий
activity_mapping = {}

# список действий в каталоге "data"
subdirectories = [subdir for subdir in os.listdir(data_directory) if os.path.isdir(os.path.join(data_directory, subdir))]

# Заполнение словаря действиями
for subdirectory in subdirectories:
    activity_mapping[subdirectory] = subdirectory

# чтение всех подкаталогой и определение действий
for subdirectory in subdirectories:
    directory_path = os.path.join(data_directory, subdirectory)
    
    # список файлов
    files = [file for file in os.listdir(directory_path) if file.endswith(".csv")]
    
    # действия для каждого файла
    for file_name in files:
        file_path = os.path.join(directory_path, file_name)
        print(f"Файл: {file_path}, Діяльність: {activity_mapping[subdirectory]}")
```
