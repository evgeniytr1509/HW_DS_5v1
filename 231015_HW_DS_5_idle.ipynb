```python
import pandas as pd
import os
import numpy as np
from scipy.stats import entropy

#  путь к каталогу, содержащему файлы CSV
directory_path = r'f:\Dell\Data_Science\HW_DS_5\data\idle'

#  пустой список, в который будут добавлены все DataFrame
all_dataframes = []

# список файлов в указанном каталоге
file_list = os.listdir(directory_path)

# чтение  каждого файла и сохранение его в DataFrame
for file_name in file_list:
    if file_name.endswith(".csv"):  # Убедитесь, что файл имеет расширение .csv
        file_path = os.path.join(directory_path, file_name)
        df = pd.read_csv(file_path)
        
        #  максимальные и минимальные значения для каждой оси движения
        max_x = df['accelerometer_X'].max()
        min_x = df['accelerometer_X'].min()
        max_y = df['accelerometer_Y'].max()
        min_y = df['accelerometer_Y'].min()
        max_z = df['accelerometer_Z'].max()
        min_z = df['accelerometer_Z'].min()
        
        # интерквартильный диапазон для столбца "accelerometer_X"
        iqr_x = df['accelerometer_X'].quantile(0.75) - df['accelerometer_X'].quantile(0.25)
        
        # Дополнительные метрики для столбца "accelerometer_Y"
        mean_y = df['accelerometer_Y'].mean()
        variance_y = df['accelerometer_Y'].var()
        std_dev_y = df['accelerometer_Y'].std()
        median_y = df['accelerometer_Y'].median()
        rms_y = np.sqrt(np.mean(df['accelerometer_Y'] ** 2))
        
        # Дополнительные метрики для столбца "accelerometer_Z"
        skewness_z = df['accelerometer_Z'].skew()
        kurtosis_z = df['accelerometer_Z'].kurt()
        iqr_z = df['accelerometer_Z'].quantile(0.75) - df['accelerometer_Z'].quantile(0.25)
        
        #  энтропию для каждой оси движения
        entropy_x = entropy(df['accelerometer_X'])
        entropy_y = entropy(df['accelerometer_Y'])
        entropy_z = entropy(df['accelerometer_Z'])
        
        # кросс-корреляцию бинарных комбинаций x, y и z
        cross_correlation_xy = np.correlate(df['accelerometer_X'], df['accelerometer_Y'], mode='same')
        cross_correlation_xz = np.correlate(df['accelerometer_X'], df['accelerometer_Z'], mode='same')
        cross_correlation_yz = np.correlate(df['accelerometer_Y'], df['accelerometer_Z'], mode='same')
        
        # все вычисленные метрики в датафрейм
        df['maximum_value_x'] = max_x
        df['minimum_value_x'] = min_x
        df['maximum_value_y'] = max_y
        df['minimum_value_y'] = min_y
        df['maximum_value_z'] = max_z
        df['minimum_value_z'] = min_z
        df['interquartile_range_x'] = iqr_x
        df['mean_y'] = mean_y
        df['variance_y'] = variance_y
        df['std_dev_y'] = std_dev_y
        df['median_y'] = median_y
        df['rms_y'] = rms_y
        df['entropy_x'] = entropy_x
        df['entropy_y'] = entropy_y
        df['entropy_z'] = entropy_z
        df['skewness_z'] = skewness_z
        df['kurtosis_z'] = kurtosis_z
        df['iqr_z'] = iqr_z
        df['cross_correlation_xy'] = cross_correlation_xy
        df['cross_correlation_xz'] = cross_correlation_xz
        df['cross_correlation_yz'] = cross_correlation_yz
        df ["id"] = 'running'
        all_dataframes.append(df)
        
# объединение в один  DataFrame:
merged_df = pd.concat(all_dataframes, ignore_index=True)
merged_df
```
